# Файла csv в DataFrame

Чтение файла csv и его переделка.   
Переделать нужно поля с датой и временем (объеденить в одно поле и сделать по нему индекс)

```python
import pandas as pd
```

```python
# Загрузка данных их файла в df (указывем конкретный файл)
df_data = pd.read_csv('../data_finam/SPFB.RTS_20200103_20200103_5min.csv')  # Загрузка данных их файла в df
df_data
```

Делаем слияние столбцов (полей) DATE и TIME. Оставляем между ними пробел для более удобного преобразования типов

```python
df_data['date_time'] = df_data['<DATE>'].astype(str) + ' ' + df_data['<TIME>'].astype(str)  # Слияние столбцов
df_data
```

```python
df_data = df_data.set_index(pd.DatetimeIndex(df_data['date_time']))  # Меняем индекс и делаем её типом date
df_data
```

```python
df_data = df_data.drop('date_time', 1)  # Удаляем ненужную колонку. 1 означает, что отбрасываем колонку а не индекс
df_data
```

```python
title_lst = df_data.columns.values  # Проверяем, что есть в заголовке
title_lst
```

В заголовке нужно избавиться от знаков "меньше" и "больше" и привести к нижнему регистру (смена регистра нужна для БД)

```python
# Удаляем ненужные символы
title_lst_new = [x.replace('<', '') for x in title_lst]  # Удаляем символ '<' в каждом элементе списка
title_lst_new = [x.replace('>', '') for x in title_lst_new]  # Удаляем символ '>' в каждом элементе списка
title_lst_new  # Отображаем список с новыми заголовками
```

```python
# Приводим список с новыми заголовками к нижнему регистру
title_lst_new = [x.lower() for x in title_lst_new]
title_lst_new
```

```python
# Создаем словарь для переименования заголовка
rename_dic = dict(zip(title_lst, title_lst_new))
rename_dic
```

```python
# Переименовываем заголовок
df_data = df_data.rename(columns = rename_dic)
df_data
```
