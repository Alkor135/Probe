# Функции, регрессия

Создадим input

```python
import matplotlib.pyplot as plt
from mysteryFunctions import mysteryFunctionOne
```

```python
# Параметры для рисования
plt.rcParams['figure.figsize'] = (16,12)
# Шрифт надписей к графикам
font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 20,
        }
```

```python
# Создадим input список
input = list(range(-10,20))
```

```python
result = mysteryFunctionOne(input)

# Центрирование оси координат
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
# Убираем верхнюю и нижнюю границу
ax.spines['right'].set_position('zero')
ax.spines['top'].set_position('zero')

plt.title('График супер секретной функции')
plt.scatter(input, result)
plt.show
```
